#### 2. Compare regional sales (North America, Europe, Japan, and Rest of the World) for each genre to understand market preferences.

```python
regional_sales_genre = df.groupby('Genre')[['North America','Europe','Japan','Rest of World']].sum()
regional_sales_genre.plot(figsize=(12, 8), kind = 'bar')
plt.title('Comparison of Regional Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```

<p>In Europe, the Action and Shooter genres lead with combined sales of around 58 million units, highlighting their strong popularity. Gamers in the region prefer fast-paced, competitive games, presenting opportunities for developers to target these genres in the European market.</p>

#### 3. Find the publisher with the highest total global sales.

#### 4. Analyze the year with the highest number of game releases.

```python
games_per_year = df.groupby('Year')['Game'].count()
games_per_year_sorted = games_per_year.sort_values(ascending=False)

highest_year = games_per_year_sorted.idxmax()
highest_count = games_per_year_sorted.max()

print(f"The year with the highest number of game releases is {highest_year} with {highest_count} releases.")
```

#### 5. Determine the most popular genre globally based on total sales.

```python
#total sales
df['Total Sales'] = df[['North America', 'Europe', 'Japan', 'Rest of World','Global']].sum(axis=1)
df['Total Sales']
```

```python
total_sales_by_genre = df.groupby('Genre')['Total Sales'].sum()
total_sales_by_genre
```

```python
total_sales_by_genre_sorted = total_sales_by_genre.sort_values(ascending=False)

most_popular_genre = total_sales_by_genre_sorted.idxmax()
most_popular_genre_sales = total_sales_by_genre_sorted.max()

print(f"The most popular genre globally is '{most_popular_genre}' with total sales of {most_popular_genre_sales:.2f}.")
```

#### 6. Identify the game with the highest sales in Japan compared to other regions.

```python
game_heighest_sale_japan = df.loc[df['Japan'].idxmax()]
print(game_heighest_sale_japan[['Game', 'Japan', 'North America', 'Europe', 'Rest of World']])
```

```python
print('''The Action genre peaked in 2015 with 33 million units sold, then dropped to 26 million in 2016, before rising again to 31 million in 2017. In contrast, the Shooter genre reached its highest sales in 2015 at 37 million, followed by a decline to 31 million in 2017. Notably, both genres saw no sales in 2018, 2019, or 2020, highlighting a significant decline in popularity during those years.''')
```

#### 7. Examine if there is a correlation between sales in North America and Europe.

```python
correlation = df['North America'].corr(df['Europe'])
correlation_rounded = round(correlation, 2)
print(f"Correlation between North America and Europe sales: {correlation_rounded}")
```

#### 8. Compare the sales growth trend for "Action" and "Shooter" genres over the years..

```python
filtered = df[df['Genre'].isin(['Action','Shooter'])]
filtered
```

```python
sales_trends = filtered.groupby(['Year', 'Genre'])['Global'].sum().reset_index()

sales_pivot = sales_trends.pivot(index='Year', columns='Genre', values='Global')

sales_pivot.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Sales Growth Trend for Action and Shooter Genres')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.legend(title='Genre')
plt.grid()
plt.show()
```

#### 9. Identify the top publisher in terms of sales for each region.

<ul>
    <li>Distribution of global sales (histogram).</li>
</ul>

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Global'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Global Sales')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

<ul>
    <li>Total sales comparison for regions (bar chart).</li>
</ul>

```python
region_sales = df[['North America', 'Europe', 'Japan', 'Rest of World']].sum()

plt.figure(figsize=(10, 6))
region_sales.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Total Sales Comparison by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales (in millions)')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```

<ul>
    <li>Share of total global sales by genre (pie chart).</li>
</ul>

```python
genre_sales = df.groupby('Genre')['Global'].sum()

plt.figure(figsize=(12, 8))
ax = genre_sales.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=90, legend=False, label= None)

plt.title('Share of Total Global Sales by Genre')

plt.legend(genre_sales.index, title="Genres", bbox_to_anchor=(1.05, 0.5), loc='center left' )

plt.tight_layout()
plt.show()
```

<p>The Action genre accounts for 2.3% of total global sales in the pie chart. While this indicates that Action is a notable genre, 
it represents only a small fraction of the overall market, suggesting that other genres likely dominate global sales.</p>

<ul>
    <li>Trend of global sales over the years (line chart).</li>
</ul>

```python
sales_trend = df.groupby('Year')['Global'].sum()

plt.figure(figsize=(10, 6))
sales_trend.plot(kind='line', marker='o', color='skyblue')
plt.title('Trend of Global Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Global Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

<ul>
    <li>Top 10 best-selling games globally (horizontal bar chart).</li>
</ul>

```python
top_10_games = df.sort_values('Global', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_10_games['Game'], top_10_games['Global'], color='skyblue', edgecolor='black')
plt.title('Top 10 Best-Selling Games Globally')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Game')
plt.tight_layout()
plt.show()
```

<ul>
 <li>Correlation between regional sales (heatmap).</li>
</ul>

```python
regional_sales = df[['North America', 'Europe', 'Japan', 'Rest of World']]

correlation_matrix = regional_sales.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Spectral', fmt='.2f', linewidths=0.5)
plt.title('Correlation Between Regional Sales')
plt.tight_layout()
plt.show()
```

<ul>
<li>Contribution of each genre to regional sales (stacked bar chart).</li>
</ul>

```python
regional_sales_by_genre = df.groupby('Genre')[['North America', 'Europe', 'Japan', 'Rest of World']].sum()

regional_sales_by_genre.plot(kind='bar', stacked=True, figsize=(12, 8), cmap='Spectral')

plt.title('Contribution of Each Genre to Regional Sales')
plt.xlabel('Genre')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
